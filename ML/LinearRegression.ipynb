```python
#import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0,100,100)
noice = np.random.normal(1,10,100)
y = 3+3*x+noice
```

```python
plt.scatter(x,y)
```

```python
# mse formula is 1/2n*(actual-pred)^2
# mse = 1/n*sum(actual-pred)
# mse for linear regression will be -2/n*sum(x*(actual-pred))
# gradient descent will be m -= lr*grad
```

```python
def LinearRegression(x, y, lr, epochs):
    m = 0
    c = 0

    for _ in range(epochs):
        pred = m*x+c
        error = y-pred
        w_gradient = -(2/len(x))* np.sum(x*error)
        c_gradient = -(2/len(x))*np.sum(error)
        w_upd = lr*w_gradient
        c_upd = lr * c_gradient
        if (np.any(np.isinf(w_upd))) or (np.any(np.isinf(c_upd))) or (np.any(np.isnan(w_upd))) or (np.any(np.isnan(c_upd))):
            break
        m -= w_upd
        c -= c_upd
    f_pred = m * x + c
    return m,c,f_pred
```

```python
x_scaled = (x-np.mean(x))/np.std(x)
```

```python
m,c,pred = LinearRegression(x_scaled,y,0.001,2000)
```

```python
plt.scatter(x_scaled,y)
plt.plot(x_scaled,pred,color='r')
```
