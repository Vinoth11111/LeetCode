```python
import math
from collections import defaultdict

class MultinomialNB:
    def fit(self, X, y):
        self.classes = set(y)
        self.word_count = defaultdict(lambda: defaultdict(int))
        self.class_count = defaultdict(int)
        self.vocab = set()

        for doc, label in zip(X, y):
            self.class_count[label] += 1
            for word in doc:
                self.word_count[label][word] += 1
                self.vocab.add(word)

        self.total_docs = len(y)

    def predict(self, X):
        predictions = []
        for doc in X:
            scores = {}
            for c in self.classes:
                log_prob = math.log(self.class_count[c] / self.total_docs)
                total_words = sum(self.word_count[c].values())
                for word in doc:
                    count = self.word_count[c][word]
                    prob = (count + 1) / (total_words + len(self.vocab))
                    log_prob += math.log(prob)
                scores[c] = log_prob
            predictions.append(max(scores, key=scores.get))
        return predictions
```
