```python
import re
def data_process(text):
    # Rs Rs. or INR space optional 1-2 digit,2-3digit,1-3digit dot 1-2 digit
    #?: is used for caturing and continueing the catpure if it not used it will immedidatly return the initally captured txt(Rs)
    #log.info('Processing data to mask sensitive information and extract amount')
    pattern = '(?:Rs\.?|INR)\s?\d{1,4}(?:,\d{2})*(?:,\d{3})*(?:\.\d+)?'
    masked_sms = re.sub(r'/d{9,18}','Masked_Number',text)
    #log.info('Masked sensitive information in the text and extracted amount')
    amt = re.search(pattern,text)
    print(amt)
    if amt:
        str_amt = amt.group()
        pro_str = re.sub(r',','',str_amt)
        int_amt = pro_str[3:].strip()

        return float(int_amt),masked_sms
    else:
        return None
```

```python
val = data_process("Your Uber ride fare Rs. 2100.10 paid through HDFC UPI on 18-09-2023. Ref ID: UBR987.")
print(val)
```

```python
def sol(nums):
    pref_sum = [0]
    suf_sum = [0]
    suf_ind = -1
    for i in range(len(nums)):
        if i == 0:
            continue
        else:
            pref_sum.append(pref_sum[i-1]+nums[i-1])
        if i == 0:
            continue
        else:
            suf_sum.append(suf_sum[i-1]+nums[suf_ind])
        count = 0
        for pre,suf in zip(pref_sum,suf_sum[::-1]):
            if pre == suf:
                return count
            else:
                count+=1
        suf_ind += -1
    return -1
```

```python
print(sol([7,3,1,3,5,6]))
print(sol([2,-1,1]))
print(sol([1,2,3]))
```

```python
def pivotIndex(nums):
        prefix_sum = [0]
        sufix_sum = [0]
        suf_ind = -1
        for i in range(len(nums)):
            if i == 0:
                continue
            else:
                prefix_sum.append(prefix_sum[i-1] + nums[i-1])
            if i == 0:
                continue
            else:
                sufix_sum.append(sufix_sum[i-1]+nums[suf_ind])
            suf_ind += -1
        count = 0
        for pre,suf in zip(prefix_sum,sufix_sum[::-1]):
            if pre == suf:
                return count
            else:
                count += 1
        return -1
```

```python
print(pivotIndex([7,3,1,3,5,6]))
print(pivotIndex([2,-1,1]))
print(pivotIndex([1,2,3]))
```

```python
from collections import defaultdict
def uniqueOccurrences(arr):
        occurences = defaultdict(int)
        for elem in arr:
            occurences[elem] += 1

        values = set(occurences.values())
        return len(values) == len(occurences.values())
```

```python
from typing import List

class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        if not matrix or not matrix[0]:
            return []

        res = []

        top = 0
        bottom = len(matrix) - 1
        left = 0
        right = len(matrix[0]) - 1

        while top <= bottom and left <= right:

            # 1️⃣ left → right (top row)
            for col in range(left, right + 1):
                res.append(matrix[top][col])
            top += 1   # shrink top boundary

            # 2️⃣ top → bottom (right column)
            for row in range(top, bottom + 1):
                res.append(matrix[row][right])
            right -= 1  # shrink right boundary

            # 3️⃣ right → left (bottom row)
            if top <= bottom:
                for col in range(right, left - 1, -1):
                    res.append(matrix[bottom][col])
                bottom -= 1

            # 4️⃣ bottom → top (left column)
            if left <= right:
                for row in range(bottom, top - 1, -1):
                    res.append(matrix[row][left])
                left += 1

        return res
```

```python
def lengthOfLongestSubstring(s: str) -> int:
    # Dictionary to store the most recent index of each character
    char_map = {}
    start = 0  # Start of the sliding window
    max_length = 0  # Maximum length of substring without duplicates

    # Loop through the string with 'end' as the end of the sliding window
    for end in range(len(s)):
        # If the character is already in the window, move 'start' pointer
        if s[end] in char_map and char_map[s[end]] >= start:
            start = char_map[s[end]] + 1

        # Update the most recent index of the current character
        char_map[s[end]] = end

        # Calculate the length of the current window and update max_length
        max_length = max(max_length, end - start + 1)

    return max_length
```
