```python
import re
```

```python
# Extract numbers only
text = "Order ID: 4598, Quantity: 23, Price: 560"
res = re.findall(r'\d',text)
print(res)
# extract Number wholely
res = re.findall(r'\d{2,4}',text)
print(res)
```

```python
# Extract all words only
text = "Python is amazing, and regex is powerful!"
res = re.findall(r'\w+',text)
print(res)
```

```python
# Find Words Starting with 'p' or 'P'
text = "Python pandas numpy pyspark PowerBI"
res = re.findall(r'[P]\w+|[p]\w+',text)
print(res)
```

```python
# Count How Many Numbers Exist
text = "There are 3 apples, 12 bananas and 5 mangoes"
res = re.findall(r'\d+',text)
print(len(res))
```

```python
# extract full numbers
text = "I bought 12 apples, 300 oranges and 5 bananas"
res = re.findall(r'\d+',text)
print(res)
```

```python
# Only extract words having 4 or more characters.
text = "ML is cool but machine learning is powerful"
#res = re.findall(r'\b\w{4,}\b',text)
res = re.findall(r'\b[a-zA-Z]{4,}\b',text)
print(res)
```

```python
# Extract Valid Years (1900â€“2099)
text = "India got independence in 1947, AI boom in 2023, future 2100"
res = re.findall(r'\b[1-2][^1-8][0-9][0-9]\b',text)
print(res)
```

```python
# only the username, not domain.
text = "Emails: john_doe@gmail.com, admin@company.org, test.user@yahoo.in"
# res = re.findall(r'\b([a-zA-Z_.]+)@[a-zA-Z.]+\b',text)
res = re.findall(r'[a-zA-Z_.]+(?=@)',text)
print(res)
```

```python
# Extract only valid phone numbers.
text = "Contacts: 9876543210, 1234567890, 9123456789, 998877665"
res = re.findall(r'\b[6-9]\d{9}\b',text)
print(res)
```

```python
# Validate Email Addresses
text = """
valid: john_doe@gmail.com
invalid: john@gmail
valid: admin@company.org
invalid: test@.com
invalid: @gmail.com
"""
res = re.findall(r'[A-Za-z._]+\@[A-Za-z]+\.[a-z]+',text)
print(res)
```

```python
# Validate URLs (http / https only)
text = "Visit https://google.com, http://example.org, ftp://files.com"
res = re.findall(r'\bhttps?\b\:\/{2}[a-z]+\.[a-z]+',text,re.MULTILINE)
print(res)
```

```python
# Extract Valid Dates (DD-MM-YYYY)
text = "Dates: 12-05-2023, 31-13-2020, 01-01-1999"
#res = re.findall(r'\b(0[1-9]|30|31|[12][0-9])\-([1-9]|1[0-2])\-\d{4}\b',text)
res = re.findall(r'\b(?:0[1-9]|[12][0-9]|30|31)-(?:0[1-9]|1[0-2])-\d{4}\b', text)
print(res)
```

```python
# Extract Hashtags, _ only.
text = "Trending: #AI #machine_learning #2026 #deep-learning"
res = re.findall(r'\#\b\w+\b',text)
print(res)
```

```python
# Validate Strong Passwords
text = "Passwords: Hello123, weakpass, STRONG123, ValidPass9"
res = re.findall(r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d).{8,}$',text,re.MULTILINE)
print(res)
```
