```python
import numpy as np
#import matplotlib.pylab as plt
```

```python
def LogisticRegression(x, y, lr, epochs):
    def sigmoid(z):
        return 1/(1+(np.exp(-z)))
    m = np.zeros(x.shape[1])
    c = 0
    for _ in range(epochs):
        z = np.dot(x,m)+c
        pred = sigmoid(z)
        error = pred - y # error is always pred - actual, if we use actual - pred then we have to use - sign in gw and gb.
        print(error)

        gw = 1/len(x)*(np.dot(x.T,error)) # we dont have to use sum because dot do that already, if we use x.T * error then we have to do sum.
        gb = 1/len(x)*np.sum(error)
        if np.any(np.isnan(gw)) or np.any(np.isnan(gb)) or np.any(np.isinf(gw)) or np.any(np.isinf(gb)): #for handling nan and infinte error
            break
        m-=lr*gw
        c-=lr*gb
    return m,c,pred
```

```python
X = np.array([[0.1, 0.2], [0.4, 0.5], [0.7, 0.8]])  # Feature matrix (3 samples, 2 features)
y = np.array([0, 1, 1])
```

```python
m,c,pred = LogisticRegression(X,y,0.01,100)
```
