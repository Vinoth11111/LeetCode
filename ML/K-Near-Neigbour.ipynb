```python
import numpy as np
```

```python
def KNN(x_train,y_train,x_test,k):
    def euclidean_distance(a,b):
        return np.sqrt(((b[0]-a[0])**2) + ((b[1]-a[1])**2)) # if we have more features than (a1-b1)^2 +(a2-b2)^2 etc
    #     return np.sqrt(np.sum((a - b) ** 2)) can works for number of features.

    # (distance,label)
    res = []
    for xtest in x_test: # each testing points must be compared with training
        distance_label = []
        for xt,lab in zip(x_train,y_train):
            distance_label.append((euclidean_distance(xt,xtest),lab))
        distance_label.sort(key=lambda x:x[0]) # sort by distance
        if k < len(x_train):
            top = distance_label[:k]
        else:
            print('k is greater than the datapoints')
            break
        label_count = {}
        for i in top:
            label_count[i[1]] = label_count.get(i[1],0) + 1
        res.append(max(label_count,key= lambda x: label_count[x]))
    return res
```

```python
X_train = [
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 5],
    [7, 7],
    [8, 6]
]

y_train = [
    "A",
    "A",
    "A",
    "B",
    "B",
    "B"
]
```

```python
X_test = [
    [2, 2],
    [7, 6]
]
```

```python
label = KNN(X_train,y_train,X_test,2)
```

```python
print('predicted neighbour is ',label)
```
